```python
trial=df.copy()
trial["vol * closed price"]=trial["Close Price"]*trial["Total Traded Quantity"]
```

```python
trial=trial.set_index("Date")
```

## VWAP

```python
# MOnth & year wise VWAP

group_M = trial.groupby(pd.Grouper(freq="M"))

month_vol=list(group_M["Total Traded Quantity"].agg(np.sum))
month_mult=list(group_M["vol * closed price"].agg(np.sum))
month=[]
for i in range(len(month_vol)) :
    month.append(round(month_mult[i]/month_vol[i],2))
    
group_Y = trial.groupby(pd.Grouper(freq="Y"))

year_vol=list(group_Y["Total Traded Quantity"].agg(np.sum))
year_mult=list(group_Y["vol * closed price"].agg(np.sum))
year=[]
for i in range(len(year_vol)) :
    year.append(round(year_mult[i]/year_vol[i],2))
```

## Average price over last n days

```python
n = int(input(" Enter the value of n : "))
        
def  avg_and_prof(n)     :
        req_df=df.tail(n)

        prof=100*(req_df.iloc[0]["Close Price"]-req_df.iloc[-1]["Close Price"])/req_df.iloc[0]["Close Price"]

        print("Average price over last",n,"days is : {} ".format(round(req_df["Average Price"].mean(),2)))
        if prof >0 :      
            print("Profit : {} %".format(round(prof,2)))
        else :
            print("Loss : {} %".format(round(prof,2)))
        
            
avg_and_prof(n)
```

```python
print("1 week :",avg_and_prof(7),".........")
print("2 weeks :\n",avg_and_prof(14))
print("1 month :\n",avg_and_prof(30))
print("3 months :\n",avg_and_prof(90))
print("6 months :\n",avg_and_prof(180))
print("1 year :\n",avg_and_prof(365))
```

```python
df["Day_Perc_Change"]=df["Close Price"].pct_change()
df["Day_Perc_Change"].fillna(0 ,inplace=True)
```

```python
def trnd(k) :
    if -0.5<=k<=0.5 :
        return "Slight or no change"
    elif 0.5<k<=1 :
        return   "Slight positive"    
    elif -1<=k<-0.5 :
        return "Slight negative"        
    elif 1<=k<=3 :
        return "Positive"
    elif -3<=k<=-1 :
        return "Negative"
    elif 3<=k<=7 :
        return "Among top gainers"
    elif -7<=k<=-3 :
        return "Among top losers"
    elif k>=7 :
        return "Bull run"
    elif k<=-7 :
        return "Bull drop"
    else :
        return np.nan
    
```

```python
arr=[trnd(x) for x in list(df["Day_Perc_Change"])]
df["trend"]=arr
```
